```python
import pandas as pd
df=pd.read_csv("/content/mushroom.csv")
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
# Check data types to identify categorical columns
print(df.dtypes)

# Convert categorical columns to numerical using Label Encoding
from sklearn.preprocessing import LabelEncoder

# Initialize the encoder
label_encoder = LabelEncoder()

# Apply the encoder to all categorical columns
for column in df.columns:
    if df[column].dtype == 'object':
        df[column] = label_encoder.fit_transform(df[column])
```

```python

x=df.drop(['class'],axis=1)
x
```

```python
y=df['class']
y
```

```python
from sklearn.model_selection import train_test_split

# Split into training and testing data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVC

# Initialize the SVM model
svm_model = SVC()

# Train the model
svm_model.fit(x_train, y_train)
```

```python
from sklearn.metrics import classification_report, accuracy_score

# Predict on the test set
y_pred = svm_model.predict(x_test)
```

```python
y_pred
```

```python
classification=classification_report(y_test,y_pred)
print(classification)
```

```python
acuuracy=accuracy_score(y_test,y_pred)
acuuracy
```

```python
x = df[['cap_shape', 'cap_surface']]  # Change as needed
y = df['class']

# Encode the target variable
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y_encoded, test_size=0.2, random_state=42)

# Train the SVM model
svm_model = SVC(kernel='linear')  # Using a linear kernel for visualization
svm_model.fit(x_train, y_train)
```

```python
# Create a mesh grid for plotting the decision boundary
import numpy as np

h = .02  # Step size in the mesh
x_min, x_max = x_train.iloc[:, 0].min() - 1, x_train.iloc[:, 0].max() + 1
y_min, y_max = x_train.iloc[:, 1].min() - 1, x_train.iloc[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Predict over the grid
Z = svm_model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundary
plt.contourf(xx, yy, Z, alpha=0.8)

# Plot the training points
plt.scatter(x_train.iloc[:, 0], x_train.iloc[:, 1], c=y_train, edgecolors='k', marker='o', s=100, label='Train')

# Plot the test points
plt.scatter(x_test.iloc[:, 0], x_test.iloc[:, 1], c=y_test, edgecolors='k', marker='x', s=100, label='Test')

# Adding labels and title
plt.xlabel('Cap Shape')
plt.ylabel('Cap Surface')
plt.title('SVM Decision Boundary')

# Show legend
plt.legend()

# Show plot
plt.show()
```
